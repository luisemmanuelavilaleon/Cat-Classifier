```python
def rename_files(path, label_name):
    fpath = main_path + path + '/'
    files = os.listdir(fpath)
    len_folder = len(files)
    for i in range(len_folder):
        os.rename(fpath + files[i], fpath + f'{label_name}_{i}.jpg')
```

```python
for label in labels:
    os.makedirs('../dataset/' + label + '/')
```

```python
def resize_files(path, label_name):
    dataset = '../dataset/'
    fpath = main_path + path + '/'
    files = os.listdir(fpath)
    len_folder = len(files)
    for i in range(len_folder):
        file = fpath + files[i]
        img = cv2.imread(file)
        img = cv2.resize(img, (256, 256))
        cv2.imwrite(dataset + label_name + '/' + f'{label_name}_{i}.jpg', img)
    print(f'{label_name} done!')
```

```python
for i in range(len(labels)):
    resize_files(folders[i],labels[i])
```

### Creating a train-test-val folders

```python
files = [os.listdir('../dataset/' + label) for label in os.listdir('../dataset/')]
```

```python
os.makedirs('../train/')
os.makedirs('../test/')
os.makedirs('../valid/')
```

```python
for i, data_file in enumerate(files):
    data_size = len(data_file)
    test_size = int(len(data_file) * 0.2)
    train_size = data_size - test_size
    test = random.sample(data_file, test_size)
    train = list(set(data_file) - set(test))
    val_size = int(train_size * 0.2)
    val = random.sample(train, val_size)
    train = list(set(train) - set(val))
    mainpath = '../dataset/' + labels[i] + '/'
    for i in range(len(train)):
        img = cv2.imread(mainpath + train[i])
        cv2.imwrite('../train/' + train[i], img)
    for i in range(len(test)):
        img = cv2.imread(mainpath + test[i])
        cv2.imwrite('../test/' + test[i], img)
    for i in range(len(val)):
        img = cv2.imread(mainpath + val[i])
        cv2.imwrite('../valid/' + val[i], img)
```

```python
label_dict = {
    'angora' : 0,
    'bengali' : 1,
    'mainecoon': 2,
    'persian' : 3,
    'russianblue' : 4,
    'siamese' : 5,
    'siberian' : 6
}
```

```python
train_set = os.listdir('../train/')
test_set = os.listdir('../test/')
valid_set = os.listdir('../valid/')
```

```python
train_labels = [label_dict[train_set[i].split('_')[0]] for i in range(len(train_set))]
test_labels = [label_dict[test_set[i].split('_')[0]] for i in range(len(test_set))]
valid_labels = [label_dict[valid_set[i].split('_')[0]] for i in range(len(valid_set))]
```

```python
train_dataset = pd.DataFrame({
    'image' : train_set,
    'label' : train_labels
})

test_dataset = pd.DataFrame({
    'image' : test_set,
    'label' : test_labels
})

valid_dataset = pd.DataFrame({
    'image' : valid_set,
    'label' : valid_labels
})
```

```python
train_dataset.to_csv('../train/train.csv', index=False)
test_dataset.to_csv('../test/test.csv', index=False)
valid_dataset.to_csv('../valid/valid.csv', index = False)
```
